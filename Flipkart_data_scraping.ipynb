```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome("/Users/yashwantjangid/Documents/chromedriver")
```

```python
driver.get("https://www.flipkart.com/search?q=smart+phones&sid=tyy%252C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_0_8&otracker1=AS_QueryStore_OrganicAutoSuggest_0_8&as-pos=0&as-type=RECENT&as-backfill=on")
```

```python
content = driver.page_source
# parse the html using beautiful soup and store in variable `soup`
soup = BeautifulSoup(content)
```

```python
products=[] #List to store name of the product
SellingPrices=[] #List to SellingPrices of the product
OriginalPrice=[]  #List to OriginalPrice of the product=[]
Product_Desc=[] #List to Product Description
Specifiactions= []# List to Product Specification
ratings=[] 
```

<font color=blue>Get List of all the links of Mobiles  <br></font>

```python
links=[] #List to all product links
for a in soup.findAll('a',href=True, attrs={'class':'_31qSD5'}):
    
    links.append(a.get('href'))
```

```python
import requests
from urllib.request import urlopen
import re
```

```python
products=[] #List to store name of the product
SellingPrices=[] #List to SellingPrices of the product
OriginalPrice=[]  #List to OriginalPrice of the product=[]
Product_Desc=[] #List to Product Description
Specifiactions= []# List to Product Specification
ratings=[] #List to store rating of the product
Description=[]# List to descriptions of product
imageUrls=[]
for l in links[:18]:
    link = "https://www.flipkart.com"+l
    driver.get(link)
    content1 = driver.page_source
    soup = BeautifulSoup(content1)
    #page = requests.get("https://www.flipkart.com"+l)
    #pagesoup = BeautifulSoup(page.text, 'html.parser')
    nameofpro = soup.find('span',attrs={'class':'_35KyD6'}).getText()
    SPrices=soup.find('div', attrs={'class':'_1vC4OE _3qQ9m1'}).getText()
    OPrice=soup.find('div', attrs={'class':'_3auQ3N _1POkHg'}).getText()
    
    Desc=soup.find('div',attrs={'class':'_3la3Fn _1zZOAc'}).getText()
    images = soup.find('div',attrs={'_2_AcLJ'})
   # for image in images:
       # imageUrls.append(image['src'])
    tags = soup.body.find_all('div',attrs={'class':'_2_AcLJ','style':True})

    spec=soup.find('div',attrs={'class':'_3Rrcbo'}).getText()
    
    rating=soup.find('div', attrs={'class':'hGSR34'}).getText()
    products.append(nameofpro)
    SellingPrices.append(SPrices)
    OriginalPrice.append(OPrice)
    Description.append(Desc)
    Specifiactions.append(spec)
    ratings.append(rating) 
    imageUrls.append(images['style'])
    
     
        
        

    
```

```python
df = pd.DataFrame({'Product Name':pd.Series(products),
                   'Original Price':pd.Series(OriginalPrice),
                   'Average Rating':pd.Series(ratings),
                   'Selling Price':pd.Series(SellingPrices),
                   'Specifiactions':pd.Series(Specifiactions),
                   'Description':pd.Series(Description),
                    'Product Link':pd.Series(links),
                   'Image Urls':pd.Series(imageUrls)
                  }) 
df.style.set_properties(subset=['Specifiactions'], **{'width': '100px'})

df.to_csv('Products1.csv', index=False, encoding='utf-8')
```
